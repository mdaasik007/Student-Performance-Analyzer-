Student Performance Analyzer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 🔹 Step 1: CSV File Creation
# 🔹 Step 2: Load with Pandas
df = pd.read_csv("student_marks_practice.csv")


# 🔹 Step 3: New Columns

# Creating Total column
df["Total"] = df["Maths"] + df["Science"] + df["English"]
# Average column
df["Average"] = df["Total"] / 3

# Grade column
df["Grade"] = np.where(df["Average"] >= 90, "A",
                       np.where(df["Average"] >= 80, "B",
                                np.where(df["Average"] <= 70, "C", "D")))

#Check Pass or fail
df["Result"] = "Fail"
df.loc[df["Average"] >= 40, "Result"] = "Pass"

#🔹 Step 4: NumPy Analysis
# 1. Convert Pandas column to NumPy array : using .values()
marks = df["Average"].values  # Pandas → NumPy array

# 2. Class Average → np.mean()
class_avg = np.mean(marks)
print("Class Average:", class_avg)

# 3. Highest Mark → np.max()
highest_mark = np.max(marks)
print("Highest Mark:", highest_mark)

# Lowest Mark → np.min()
lowest_mark = np.min(marks)
print("Lowest Mark:", np.min(marks))

# Top 5 Students → np.sort()
top5 = np.sort(marks)[-5:]  # last 5 values = top 5
print("Top 5 Students:", top5)

# 6. Pass Percentage
pass_count = np.sum(marks >= 40) # count of pass
total_students = len(marks)
pass_percentage = (pass_count / total_students) * 100
print("Pass Percentage:", pass_percentage)


# 🔹 Step 5: Matplotlib Visualization
# Histogram plot
bins = [0, 20, 40, 60, 80, 100]
plt.hist(marks, bins=bins, edgecolor="black")
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel(" Number of Students")
plt.show()
print()
# 🟢 Pie Chart – Pass vs Fail

#check pass and fail count
pass_count = np.sum(df["Result"] == "Pass")
fail_count = np.sum(df["Result"] == "Fail")

# 2. Prepare Data
values = [pass_count , fail_count]
labels = ["Pass", "Fail"]
colors = ["green", "red"]

# 3. Plot Pie Chart
plt.pie(values, labels =labels, colors = colors, autopct="%1.1f%%", startangle=90)
plt.title("Pass or Fail Distribution")
plt.show()
print()

# 2. Subject‑wise Pie Chart (Optional Extra Practice)
maths_pass = np.sum(df["Maths"] >= 40)
maths_fail = np.sum(df["Maths"] < 40)

plt.pie([maths_pass,maths_fail],
        labels = ["Pass", "Fail"],
        colors = ["blue", "orange"],
        autopct="%1.1f%%",
        startangle=90)
plt.title("Maths Pass vs Fail")
plt.show()
```
